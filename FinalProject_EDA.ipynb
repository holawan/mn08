### 연구 문제 
 **미네소타 주 내 각 카운티**(87 지역)의 <code>2008년 미국 대통령 선거 결과</code>(오바마가 우세)를 바탕으로 **두 주요 후보**(민주당의 버락 오바마, 공화당의 존 메케인)에 중점을 두고 카운티에서 어떤 공변량으로 인해 <code>오바마가 매케인보다 더 많은 표를 얻었는지</code> 분석하고자 한다.  
 
 당시 미네소타 주의 승자는 <code>오바마</code> 였다.
 
 이와 관련하여 지역적 특성 및 두 대선 후보의 정책의 차이에 대해 선행 자료를 통해 검토해볼 필요가 있다. 
 
 **미국은 선거인단 제도를 채택하여, 인구비례로 주마다 선거인단을 뽑는다. 실제 대통령 및 부통령에게 투표를 하는 것은 선거인단이기 때문에, 대부분 유권자들(시민)의 의견과 일치하지만 그렇지 않은 경우도 존재한다.  
 또한, 각 주의 선거인단의 투표 후 과반을 차지한 후보가 해당 주의 선거인단의 표를 모두 가져간다. (승자독식)**
- <code>매케인(공화당)</code> 
     - 전반적 감세정책을 통한 기업활동 활성화
     - 미국 근로자들의 국제경쟁력 제고
     - 정부지출 자제를 통한 재정적자 해소
     - 일자리 창출 통한 고용불안 해소
- <code>오바마(민주당)</code> 
     - 증세, 임금인상, 공공서비스 확충을 통한 부의 불균형 해소
     - 미국산업 및 근로자들의 일자리 보호
     - 기업 및 고소득층 증세를 통한 재정적자 해소
     - 최저임금 인상
     - 중상층 및 저소득층 지원통한 소비위축 완화
 
 

미네소타주에 대해 조사한 결과, (현재) 미국 중서부 지역에서 가장 큰 주이며 주로 서비스업, 제조업, 농업에 종사하는 것으로 나타난다.
 
당시 상황을 살펴보면, <code>2008년 당시 미국은 금융위기</code>를 맞았고, 대선에서 각 주요 후보는 거의 상반된 입장을 취했다는 것을 알 수 있다. 

미네소타주는 2008년 이전 10년여간 민주당이 승리를 9회 차지했으며, <code>8년 연속 (1976년~2004년) 민주당이 승리</code>를 차지했다.

자료의 형태를 살펴보면, **수입, 실업, 시골 거주 여부**등의 변수들이 주를 이루어, 이와 관련된 <code>경제정책의 차이</code>가 미네소타 주 내 카운티의 투표결과에 영향을 끼쳤는지 검증해보고자 한다

### 자료 소개

**mn08.csv의 자료는 미네소타 주 카운티의 2008년 미국 대통령 선거 결과를 포함하고 있다: 두 주요 후보 (민주당 바락 오바마와 공화당 존 매케인)에 중점을 두고 있다. 카운티에서 오바마 투표의 비율 또는 오바마가 McCain 보다 더 많은 표를 얻었는지 여부를 반응변수로 고려할 수 있다.**  

자료는 총 12개의 변수로 구성되어 있으며, 카운티는 총 87곳이다. 

- <code>County</code> = county name (미네소타주 내 카운티의 이름)
- <code>Obama</code> = total votes for Obama (카운티 별 오바마가 받은 표의 수)
- <code>McCain</code> = total votes for McCain (카운티 별 메케인이 받은 표의 수)
- <code>pct_Obama</code> = percent of votes for Obama (오바마의 득표비율)
- <code>pct_rural</code> = percent of county who live in a rural setting (시골에 거주하는 카운티의 비율)
- <code>medHHinc</code> = median household income(수입의 중앙값)
- <code>unemp_rate</code> = unemployment rate (실업자 비율)
- <code>pct_poverty</code> = percent living below the poverty line (저소득층의 비율)
- <code>medAge2007</code> = median age in 2007 (2007년 당시 나이의 중위수)
- <code>medAge2000</code> = median age in 2000 (2000년 당시 나이의 중위수)
- <code>Gini_Index</code> = measure of income disparity in a county (카운티 내의 소득 격차)
- <code>pct_native</code> = percent of native born residents (토착주민의 비율)


38곳의 카운티는 medAge2007, Gini_index, pct_native가 모두 입력되지 않았다.  
Lincoin, Marshall의 경우에는 pct_native만 입력되지 않았고, Martin, Mile Lacs의 경우에는 medAge2007과, Gini_Index가 입력되지 않았다.  
따라서 결측치가 존재하는 3개의 변수 (medAge2007, Gini_index, pct_native)가 반응변수에 영향을 미치는 공변량이라면 해당 카운티를 모두 제거할 것이고, 영향이 없다면 추가해도 될 것이라고 생각한다.

    - 오바마가 과반을 차지한 지역은 저소득층의 비율은 평균적으로 높았으며, 소득의 격차도 더 큰 차이를 보였다. (오바마의 저소득층 정책의 영향?)
    - 토착주민의 비율이 평균적으로 높은 곳에서는 매케인이 과반을 가져갔음을 알 수 있다.
    
- pct_rural
    - 시골에 거주하는 카운티의 비율은 수입의 중앙값과 음의 상관관계를 보였고 실업자 비율, 저소득층 비율, 토착주민 비율과 양의 상관관계를 보인다.
    - 지니계수와는 상관이 없는 것으로 나타난다.
    - 시골에 거주하는 비율이 높을수록 수입이 적고, 실업자, 저소득층, 토착주민의 비율이 많음을 알 수 있다.

- medHHinc
    - 수입의 중앙값은 전체적으로 모든 변수와 음의 상관관계를 보인다.
    - 시골거주 비율, 토착주민 비율, 실업자, 저소득층 비율, 소득격차가 있으며 그 중 저소득층 비율과는 큰 음의 상관관계를 보인다.
    - 이 지역들은 도시가 아닐까?

- unemp_rate
    - 실업자 비율과 시골거주 비율 & 토착주민의 비율은 양의 상관관계를 보인다. 소득격차와는 상관이 없는 것으로 보이며, 저소득층 비율과도 양의 상관관계를 보인다.
    - 수입의 중앙값과는 낮게 나타난다.
    - 시골에 거주하는 비율, 토착주민의 비율과 모두 실업자와 강한 양의 상관관계를 보이는 것으로 미루어보아, 시골의 실업률이 높음을 알 수 있다.

- pct_poverty
    - 저소득층의 비율은 지니계수와 강한 양의 상관관계를 보이며, 수입의 중앙값과는 강한 음의 상관관계를 보인다. 또한 시골 거주 비율과도 양의 상관관계가 있다.
    - 따라서, 저소득층의 비율이 높은 곳은 소득격차 또한 많이 발생하며, 평균적으로 수입이 적다는 것을 알 수 있다.

- Gini_Index
    - 소득격차는 수입의 중앙값&토착주민의 비율과는 음의 상관관계를 보이고, 저소득층의 비율, 오바마 득표율과는 높은 양의 상관관계를 보인다.
    - 오바마의 부의 불균형 해소 정책이 영향이 있는 것 같다.
- pct_Obama
    - 오바마의 득표율은 지니계수와 양의 상관관계를 보이며, 토착주민 비율과는 음의 상관관계를 보인다.
    - 민주당이 8년 이상 집권을 했다는 것이 놀랍다.
- Num
    - 투표수를 인구라고 가정했을 때, 인구가 많은 지역은 수입의 중앙값과, 오바마 득표율에 양의 상관관계가 있었다.
    - 시골 거주 비율과 토착 주민 비율과는 음의 상관관계를 보인다.
